### Comparación de Crecimiento en los ultimos 82 días

---



Oyentes

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

time = ['13074', '25244']
counts = [13074, 25244]
bar_labels = ['23 Mar - 13 Jun', '14 Jun - 04 Sep']
bar_colors = ['tab:red', 'tab:blue']

ax.bar(time, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('Cantidad de Oyentes')
ax.set_title('Crecimiento de Oyentes Spotify')
ax.legend(title='Periodo')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos originales
time = ['23 Mar - 13 Jun', '14 Jun - 04 Sep']
counts = [13074, 25244]

# Calcular el porcentaje de crecimiento
total_count = sum(counts)
growth_percentages = [(count / total_count) * 100 for count in counts]

# Colores para el gráfico de pastel
pie_colors = ['tab:red', 'tab:blue']

# Crear el gráfico de pastel
fig, ax = plt.subplots()
ax.pie(growth_percentages, labels=time, autopct='%1.1f%%', colors=pie_colors, startangle=90)
ax.axis('equal')  # Aspecto de círculo

ax.set_title('Crecimiento de Oyentes Spotify por Periodo')

plt.show()
```
